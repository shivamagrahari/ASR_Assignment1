```python
def text_ground_truth(s):
    timit_df = pd.read_hdf("./features/newfolder/"+s+".hdf")
    for i in range(0,1680):
    #print(i)
        a = timit_df.loc[timit_df['ids'] == i]
        x = a['labels'].values
        fo = open("ground_truth_"+s+".txt","a+")
    #fo.write("Sentence"+" "+str(i)+"\n")
        for label in x:
            fo.write(label+" ")
        fo.write("\n")
    fo.close()
```

```python
text_ground_truth("timit_test")
text_ground_truth("timit_test_delta")
text_ground_truth("timit_test_delta_delta")
```
